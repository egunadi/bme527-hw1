Get the DICOM Data Set

```python
dcm_dataset01_filepath = '../data/dcm/image01.dcm'
dcm_dataset02_filepath = '../data/dcm/image02.dcm'
dcm_dataset03_filepath = '../data/dcm/image03.dcm'
dcm_dataset04_filepath = '../data/dcm/image04.dcm'
dcm_dataset05_filepath = '../data/dcm/image05.dcm'
dcm_dataset06_filepath = '../data/dcm/image06.dcm'
dcm_dataset07_filepath = '../data/dcm/image07.dcm'
dcm_dataset08_filepath = '../data/dcm/image08.dcm'
dcm_dataset09_filepath = '../data/dcm/image09.dcm'
dcm_dataset10_filepath = '../data/dcm/image10.dcm'
dcm_dataset11_filepath = '../data/dcm/image11.dcm'
dcm_dataset12_filepath = '../data/dcm/image12.dcm'
dcm_dataset13_filepath = '../data/dcm/image13.dcm'
dcm_dataset14_filepath = '../data/dcm/image14.dcm'
dcm_dataset15_filepath = '../data/dcm/image15.dcm'
dcm_dataset16_filepath = '../data/dcm/image16.dcm'
dcm_dataset17_filepath = '../data/dcm/image17.dcm'
dcm_dataset18_filepath = '../data/dcm/image18.dcm'
dcm_dataset19_filepath = '../data/dcm/image19.dcm'
dcm_dataset20_filepath = '../data/dcm/image20.dcm'
```

```python
dcm_dataset01 = dicom.dcmread(dcm_dataset01_filepath)
dcm_dataset02 = dicom.dcmread(dcm_dataset02_filepath)
dcm_dataset03 = dicom.dcmread(dcm_dataset03_filepath)
dcm_dataset04 = dicom.dcmread(dcm_dataset04_filepath)
dcm_dataset05 = dicom.dcmread(dcm_dataset05_filepath)
dcm_dataset06 = dicom.dcmread(dcm_dataset06_filepath)
dcm_dataset07 = dicom.dcmread(dcm_dataset07_filepath)
dcm_dataset08 = dicom.dcmread(dcm_dataset08_filepath)
dcm_dataset09 = dicom.dcmread(dcm_dataset09_filepath)
dcm_dataset10 = dicom.dcmread(dcm_dataset10_filepath)
dcm_dataset11 = dicom.dcmread(dcm_dataset11_filepath)
dcm_dataset12 = dicom.dcmread(dcm_dataset12_filepath)
dcm_dataset13 = dicom.dcmread(dcm_dataset13_filepath)
dcm_dataset14 = dicom.dcmread(dcm_dataset14_filepath)
dcm_dataset15 = dicom.dcmread(dcm_dataset15_filepath)
dcm_dataset16 = dicom.dcmread(dcm_dataset16_filepath)
dcm_dataset17 = dicom.dcmread(dcm_dataset17_filepath)
dcm_dataset18 = dicom.dcmread(dcm_dataset18_filepath)
dcm_dataset19 = dicom.dcmread(dcm_dataset19_filepath)
dcm_dataset20 = dicom.dcmread(dcm_dataset20_filepath)
```

### (b) Display Data

```python
pixel_array_01 = dcm_dataset01.pixel_array
pixel_array_02 = dcm_dataset02.pixel_array
pixel_array_03 = dcm_dataset03.pixel_array
pixel_array_04 = dcm_dataset04.pixel_array
pixel_array_05 = dcm_dataset05.pixel_array
pixel_array_06 = dcm_dataset06.pixel_array
pixel_array_07 = dcm_dataset07.pixel_array
pixel_array_08 = dcm_dataset08.pixel_array
pixel_array_09 = dcm_dataset09.pixel_array
pixel_array_10 = dcm_dataset10.pixel_array
pixel_array_11 = dcm_dataset11.pixel_array
pixel_array_12 = dcm_dataset12.pixel_array
pixel_array_13 = dcm_dataset13.pixel_array
pixel_array_14 = dcm_dataset14.pixel_array
pixel_array_15 = dcm_dataset15.pixel_array
pixel_array_16 = dcm_dataset16.pixel_array
pixel_array_17 = dcm_dataset17.pixel_array
pixel_array_18 = dcm_dataset18.pixel_array
pixel_array_19 = dcm_dataset19.pixel_array
pixel_array_20 = dcm_dataset20.pixel_array
```

```python
# show all images as a grid
fig, axs = plt.subplots(4, 5, figsize=(20, 20))
axs[0, 0].imshow(pixel_array_01, cmap='gray')
axs[0, 1].imshow(pixel_array_02, cmap='gray')
axs[0, 2].imshow(pixel_array_03, cmap='gray')
axs[0, 3].imshow(pixel_array_04, cmap='gray')
axs[0, 4].imshow(pixel_array_05, cmap='gray')
axs[1, 0].imshow(pixel_array_06, cmap='gray')
axs[1, 1].imshow(pixel_array_07, cmap='gray')
axs[1, 2].imshow(pixel_array_08, cmap='gray')
axs[1, 3].imshow(pixel_array_09, cmap='gray')
axs[1, 4].imshow(pixel_array_10, cmap='gray')
axs[2, 0].imshow(pixel_array_11, cmap='gray')
axs[2, 1].imshow(pixel_array_12, cmap='gray')
axs[2, 2].imshow(pixel_array_13, cmap='gray')
axs[2, 3].imshow(pixel_array_14, cmap='gray')
axs[2, 4].imshow(pixel_array_15, cmap='gray')
axs[3, 0].imshow(pixel_array_16, cmap='gray')
axs[3, 1].imshow(pixel_array_17, cmap='gray')
axs[3, 2].imshow(pixel_array_18, cmap='gray')
axs[3, 3].imshow(pixel_array_19, cmap='gray')
axs[3, 4].imshow(pixel_array_20, cmap='gray')
```
